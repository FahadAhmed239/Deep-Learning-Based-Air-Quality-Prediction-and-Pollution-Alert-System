st.sidebar.number_input(“CO”, 0.0, 10.0, 1.0) o3 =
st.sidebar.number_input(“O3”, 0.0, 200.0, 40.0) temp =
st.sidebar.number_input(“Temperature (°C)”, -10.0, 50.0, 25.0) hum =

st.sidebar.number_input(“Wind Speed (m/s)”, 0.0, 20.0, 3.0)
